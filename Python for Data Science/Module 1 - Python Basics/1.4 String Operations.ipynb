<h3>Slicing </h3>


Use slicing to find the first four elements of the following string:


````python
Letters="ABCDEFGHIJK"
print(Letters[0:4])
````

<h3>Strides Are Good </h3> 


Use a stride value of 2 on the following string :


````python
Good="GsoAo+d"
````
